# cattegorical plot
we can use "catplot" to categorise the data.

```
df.head(4)
```

```
sns.catplot('cut', 'price', data=df)
```

```
sns.catplot('color', 'price', data=df, jitter=False)
```

jitter -- is used to remove the magnitude and get the st. line 

```
sns.catplot('color', 'price', data=df.query("price >10000"),jitter=False )
```

it is way to filter the data 


# Swarm plot -

```
sns.catplot('color', 'price', data=df, kind="swarm")
```

```
sns.catplot('color', 'price', data=df.query("price>5000"), kind="swarm")
```

It is next level of chart which is used to find out the outliner.

```
sns.catplot('cut', 'price', data=df, kind="violin")
```

```
sns.catplot('cut', 'price', data=df, hue='clarity',col='color',col_wrap=2, kind='violin')
```

# Distribution Plot
It is way to represent the data in curve such that we can visulalize the data pattern and their distribution

```
sns.displot(df.price)
```

```
sns.displot(df.depth, kde=False)
```

```
sns.distplot(df.depth, hist=False)
```

```
sns.distplot(df.depth, hist=False, rug=True)
```

"rug" shows the conventration of data point over the region

```
sns.distplot(df.depth, hist=False, rug=True, bins=1000)
```
